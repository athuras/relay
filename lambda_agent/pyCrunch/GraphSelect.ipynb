# Optimal Behaviour-Trajectory Planning
Given a prediction of the flow of traffic expected at each inlet to an intersection,
determine how, and when to change intersection behaviour, to maximize throughput.

Assumptions:

* Incident flow at an inlet, if not served, accumulates in a queue.

* Once an inlet is being served, the queue depletes—adding flow (governed by the queue *relief* model) to the predicted value over the behaviour time period.

The Relay planner operates under the assumption that the predictions are valid, and seeks to maximize the liklyhood that it serves a stream of vehicles, rather than the discrete vehicles themselves.

What follows is a break-down of the behaviour.

## Prediction Generator

```python
def gen_prediction(inlets=4, dt=1, tmax=200, amplitudes=10, osc=4, tau=20):
    '''Generate some sinusoidal predictions, out of phase'''
    phases = [i * 1.6 * np.pi / inlets for i in xrange(inlets)]
    t = np.arange(0., tmax, dt)
    s = np.exp(-t/tau)
    amps = [amplitudes] * inlets if type(amplitudes) is not list else amplitudes
    F = np.vstack(a * np.sin(osc * np.random.rand() * np.pi / tmax * t + p) * s
                  for a, p in zip(amps, phases))
    return np.abs(F)

dt=1.
F = gen_prediction(4, 1, 240, [1, 0.8, 0.4, 1], osc=3, tau=140)
plt.plot(F.T, label=' ')
plt.legend();
```
