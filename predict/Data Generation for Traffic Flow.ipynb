# Generating Test Data for ML Prediction

1. Model underlying ditribution as Gamma/Erlang based on physical properties
2. Model noise as a collection of folded normals (with arbitrary scale)
3. Model input events as (more-or-less) periodic

```python
# Start by defining the 'Edge'
distance = 450.  # meters
speed_limit = 50 * 1000. / 3600  # m/s
driver_speed_cap = 1.2  # the fastest speed we want to allow
min_time = distance / (driver_speed_cap * speed_limit)
```

## Underlying Distribution
The thing we want to approximate (with some consideration for noise etc.)

```python
master = stats.gamma(3, loc=min_time, scale=4)

time = np.linspace(20, 120, 121)
speed = distance / time * 3.6
f, (a1, a2) = plt.subplots(1, 2, sharey=True)
a1.plot(time, master.pdf(time))
a1.set_title("Time Delay Distribution")
a2.plot(speed, master.pdf(time))
a2.set_title("Speed Distribution")
```

## Noise Modelling
Why a folded-norm? Because: thats why.

```python
n1 = stats.foldnorm(1, scale=2)
n2 = stats.foldnorm(1, loc=5, scale=1)
x2 = np.linspace(0, 10, 100)
plt.plot(x2, n1.pdf(x2) + n2.pdf(x2))
```

## Generating Events

```python
x3 = np.linspace(0, 1000, 500)
sqr = signal.square(2 * np.pi * 5 * x3)
sig = np.sin(6 * np.pi * x3)
pwm = signal.square(2 * np.pi * 50 * x3, duty=(sig + 1)/2)
```

```python
pre = np.where(pwm == 1)[0]
lag = master.rvs(size=pre.size)
noise = n1.rvs(size=pre.size) + n2.rvs(size=pre.size)
post = pre + lag + noise
plt.plot(pre - post)
```

## The Script

```python
def generate_data(distance, speed_limit, fast_coeff, amount):
    '''
    Generate two time-linked event measurements.
    * distance: In meters, the length of the edge/road
    * speed_limit: in km/h
    * fast_coeff: how much faster than the speed limit we 
        want to allow, typically 1.2
    * amount: the number of samples to generate
    '''
    min_time = distance / (fast_coeff * speed_limit / 3.6)
    G = stats.gamma(3, loc=min_time, scale=4)
    #x = np.linspace(1, 6*amount, 6*amount)
    #sqr = signal.square(5 * np.pi * x)
    #sig = np.sin(3 * np.pi * x)
    #pwm = signal.square(5 * np.pi * x, duty=(sig + 1) / 4)
    
    sample = stats.foldnorm(1, scale=3)
    
    A = sample.rvs(100)#np.where(pwm == 1)[0]
    
    # Noise
    n1 = stats.foldnorm(1, scale=2)
    n2 = stats.foldnorm(1, loc=5, scale=1)
    noise = n1.rvs(size=A.size) + n2.rvs(size=A.size)
    
    lag = G.rvs(size=A.size)
    B = A + lag + noise
    A = A + noise
    u = B.mean() - A.mean()
    s = (A.std() + B.std()) / 2 
    a = A#[(A >= u - 2 * s) & (A <= u + 2 * s)]
    b = B#[(B >= u - 2 * s) & (B <= u + 2 * s)]
    
    # minslice
    q = min(a.size, b.size, amount)
    return G, a[:q], b[:q]
```

```python
G, A, B = generate_data(1000, 60, 1.2, 400)
```
