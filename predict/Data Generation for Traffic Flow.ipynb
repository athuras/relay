```python
def generate_data(distance, speed_limit, fast_coeff, amount):
    '''
    Generate two time-linked event measurements.
    * distance: In meters, the length of the edge/road
    * speed_limit: in km/h
    * fast_coeff: how much faster than the speed limit we 
        want to allow, typically 1.2
    * amount: the number of samples to generate
    '''
    min_time = distance / (fast_coeff * speed_limit / 3.6)
    G = stats.gamma(3, loc=min_time, scale=3)
    #x = np.linspace(1, 6*amount, 6*amount)
    #sqr = signal.square(5 * np.pi * x)
    #sig = np.sin(3 * np.pi * x)
    #pwm = signal.square(5 * np.pi * x, duty=(sig + 1) / 4)
    
    sample = stats.foldnorm(1, scale=3)
    
    A = sample.rvs(100)#np.where(pwm == 1)[0]
    
    # Noise
    n1 = stats.foldnorm(1, scale=2)
    n2 = stats.foldnorm(1, loc=5, scale=1)
    noise = n1.rvs(size=A.size) + n2.rvs(size=A.size)
    
    lag = G.rvs(size=A.size)
    B = A + lag + noise
    A = A + noise
    u = B.mean() - A.mean()
    s = (A.std() + B.std()) / 2 
    a = A#[(A >= u - 2 * s) & (A <= u + 2 * s)]
    b = B#[(B >= u - 2 * s) & (B <= u + 2 * s)]
    
    # minslice
    q = min(a.size, b.size, amount)
    return G, a[:q], b[:q]
```

```python
def generate_nbr_signals(distance, speed_limit, fast_coeff, amount):
    '''
    generate set of signals
    * distance: In meters, the length of the edge/road
    * speed_limit: in km/h
    * fast_coeff: how much faster than the speed limit we 
        want to allow, typically 1.2
    * amount: the number of samples to generate
    '''
    min_time = distance / (fast_coeff * speed_limit / 3.6)
    G = stats.gamma(3, loc=min_time, scale=4)
    #x = np.linspace(1, 6*amount, 6*amount)
    #sqr = signal.square(5 * np.pi * x)
    #sig = np.sin(3 * np.pi * x)
    #pwm = signal.square(5 * np.pi * x, duty=(sig + 1) / 4)
    
    sample = stats.foldnorm(1, scale=3)
    
    A = sample.rvs(100)#np.where(pwm == 1)[0]
    
    # Noise
    n1 = stats.foldnorm(1, scale=2)
    n2 = stats.foldnorm(1, loc=5, scale=1)
    noise = n1.rvs(size=A.size) + n2.rvs(size=A.size)
    
    #lag = G.rvs(size=A.size)
    C = A + 2.2 * noise
    B = A + 4 * noise
    A = A + noise
    u = B.mean() - A.mean()
    s = (A.std() + B.std()) / 2 
    a = A#[(A >= u - 2 * s) & (A <= u + 2 * s)]
    b = B#[(B >= u - 2 * s) & (B <= u + 2 * s)]
    c = C
    
    # minslice
    q = min(a.size, b.size, c.size, amount)
    return G, min_time, a[:q], b[:q], c[:q]
```

```python
def create_bhvr_prob_mtx(bhvr_past):
        bhvr_prob_mtx = np.zeros([4,4])
        for edge in bhvr_past:
            bhvr_prob_mtx[edge[0], edge[1]] = edge[2]

        return bhvr_prob_mtx
```

```python
bhvr_sample = ((0, 1, 0.1), (0, 2, 0.6), (0, 3, 0.3), (3, 2, 1), (1, 0, 1), \
            (2, 1, 0.2), (2, 0, 0.7), (2, 3, 0.1))
nbr_bhvr_probs = create_bhvr_prob_mtx(bhvr_sample)

gamm, min_time, A, B, C = generate_nbr_signals(1000, 60, 1.2, 400)

l_pdf = np.linspace(0, 80, 100)
td = gamm.pdf(l_pdf)

connecting_nodes = np.array([1, 3]) # my inlet, nbr outlet
```

```python
# 30 is arbitrary -> depends on time range of values. want like 5 second sampling
bins = np.linspace(0, 100, 100)
nbr_signals = [np.histogram(A, bins), np.histogram(B, bins), np.histogram(C, bins)]
```

```python
def extract_probs(nodes, probs):
    '''
    eliminates the probably of the outlet node itself, only keeping ones that feed it
    '''
    outlet_probs = probs[:,nodes[1]]
    correct_probs = [v for k,v in enumerate(outlet_probs) if k != nodes[1]]
    
    return np.array(correct_probs)
```

```python
#Generate signals for each inlet
#tds2.plot(AH[1][0:-1], AH[0]*0.7)
def transform_signals(signal, prob):
    trans_sigs = []
    for i in range(len(prob)):
        trans_sigs.append((signal[i][0] * prob[i], signal[i][1][0:-1]))
    
    return trans_sigs
```

```python
newsig = transform_signals(nbr_signals, extract_probs(connecting_nodes, nbr_bhvr_probs))
```

```python
#(newsig[0][0] + newsig[1][0] + newsig[2][0])
def combine_signal_ests(signal):
    total_est = np.zeros(len(signal[0][0]))
    for i in range(len(signal)):
        total_est += signal[i][0]
    
    return (total_est, signal[0][1]) 
```

```python
combined_nbr_signals = combine_signal_ests(newsig)
combined_nbr_signals[0]
```
